```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.load("./models/same_base/xs.npy")
preds = np.load("./models/same_base/predictions.npy")
labels = np.load("./models/same_base/true_labels.npy")
paths = [f'./models/same_base/model_{i:02d}.pt' for i in range(100)]
```

```python
xs71 = np.load("./models/same_base71/xs.npy")
preds71 = np.load("./models/same_base71/predictions.npy")
labels71 = np.load("./models/same_base71/true_labels.npy")
paths71 = [f'./models/same_base71/model_{i:02d}.pt' for i in range(100)]
```

```python
xs33 = np.load("./models/same_base33/xs.npy")
preds33 = np.load("./models/same_base33/predictions.npy")
labels33 = np.load("./models/same_base33/true_labels.npy")
paths33 = [f'./models/same_base33/model_{i:02d}.pt' for i in range(100)]
```
