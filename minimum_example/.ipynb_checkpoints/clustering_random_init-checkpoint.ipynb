```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.load("./models/random_init/xs.npy")
preds = np.load("./models/random_init/predictions.npy")
labels = np.load("./models/random_init/true_labels.npy")
paths = [f'./models/random_init/model_{i:02d}.pt' for i in range(100)]
```

```python
num_models = len(paths)
```

```python
regression_errors = []

for i in range(num_models):
    regression_error = np.square(np.subtract(preds[i], labels)).mean()
    regression_errors.append(regression_error)
```

```python
fig, ax = plt.subplots()

for i in range(num_models):
    ax.plot(xs, np.abs(preds[i] - labels))
    
plt.show()
```
