```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.load("./models/same_base/xs.npy")
preds = np.load("./models/same_base/predictions.npy")
labels = np.load("./models/same_base/true_labels.npy")
paths = [f'./models/same_base/model_{i:02d}.pt' for i in range(100)]
```

```python
xs12 = np.load("./models/same_base12/xs.npy")
preds12 = np.load("./models/same_base12/predictions.npy")
labels12 = np.load("./models/same_base12/true_labels.npy")
paths12 = [f'./models/same_base12/model_{i:02d}.pt' for i in range(100)]
```
